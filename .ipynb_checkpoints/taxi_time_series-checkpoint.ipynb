# 3. Обучение

Выделим функцию создания признаков.

```python
def make_features(data, max_lag, rolling_mean_size):
    data['year'] = data.index.year
    data['month'] = data.index.month
    data['day'] = data.index.day
    data['dayofweek'] = data.index.dayofweek
    
    for lag in range(1, max_lag + 1):
        data['lag_{}'.format(lag)] = data['num_orders'].shift(lag)

    data['rolling_mean'] = data['num_orders'].shift().rolling(rolling_mean_size).mean()

```

Т.к. было бы нецелесообразно пытаться подобрать наилучшие параметры сдвига для каждой модели, подберем их для простой линейной регресии, а в дальнейшем будем отталкиваться от полученных max_lag и rolling_mean_size при переборе гиперпараметров других моделей.  
Для начала вычислим оптимальные max_lag и rolling_mean_size для LinearRegression  
  
Выделим метод calculate_linear_rmse_score который вычислит rmse по присланным данным

```python
def calculate_linear_rmse_score(data, max_lag, rolling_mean_size):
    make_features(data, max_lag, rolling_mean_size)
    train, test = train_test_split(data, shuffle=False, test_size=0.1)
    train = train.dropna()
    train_target = train["num_orders"]
    train_features = train.drop(columns=["num_orders"])

    test_target = test["num_orders"]
    test_features = test.drop(columns=["num_orders"])

    lin_reg_train = LinearRegression()
    lin_reg_train.fit(train_features, train_target)
    pred_train = lin_reg_train.predict(train_features)
    pred_test = lin_reg_train.predict(test_features)
    return mean_squared_error(pred_test, test_target) ** 0.5
```

Подберем оптимальные параметры в цикле.

```python
best_max_lag = 0
best_rolling_mean_size = 0
best_rmse = 1000
for max_lag in range(10,30,2):
    for rolling_mean_size in range(10,30,2):
        score = calculate_linear_rmse_score(df, max_lag, rolling_mean_size)
        if (best_rmse > score):
            best_rmse = score
            best_max_lag = max_lag
            best_rolling_mean_size = rolling_mean_size
```

```python
print("best_max_lag:", best_max_lag)
print("best_rolling_mean_size:", best_rolling_mean_size) 
```

Выделим тестовые и тренировочные выборки на основе полученных параметров.

```python
make_features(df, best_max_lag, best_rolling_mean_size)
train, test = train_test_split(df, shuffle=False, test_size=0.1)
train = train.dropna()
train_target = train["num_orders"]
train_features = train.drop(columns=["num_orders"])

test_target = test["num_orders"]
test_features = test.drop(columns=["num_orders"])
```

#### LinearRegression
Построим модель линейной регресии и посчитаем rmse

```python
lin_reg_train = LinearRegression()
lin_reg_train.fit(train_features, train_target)
pred_train = lin_reg_train.predict(train_features)
print("RMSE тренировочной выборки: ", mean_squared_error(pred_train, train_target) ** 0.5)
```

#### DecisionTreeRegressor
Также испытаем DecisionTreeRegressor с разными параметрами max_depth  
Используем GridSearchCV для подбора оптимальных параметров. В качестве параметров для перебора используем max_depth и min_samples_split

```python
param_grid = {'max_depth': range(2,20,2),
              'min_samples_split': range(2,100,2)}

clf = GridSearchCV(DecisionTreeRegressor(), 
                   param_grid,
                   scoring='neg_mean_squared_error',
                   cv=5 , verbose=1)
clf.fit(train_features, train_target) 
clf.best_params_, clf.best_score_
```

Обучим модель с полученными параметрами и посчитаем RMSE

```python
dt_regressor = DecisionTreeRegressor(max_depth = 6, min_samples_split = 82)
dt_regressor.fit(train_features, train_target)
y_pred_dt = dt_regressor.predict(train_features)
print("RMSE тренировочной выборки: ", mean_squared_error(y_pred_dt, train_target) ** 0.5)
```

#### LightGBM
Схожим образом подберем оптимальные гиперпараметры для LightGBM. Для подбора используем параметры max_bin

```python
param_grid = {
    'max_bin': range(200,1000,400)
    }

train_data = lightgbm.Dataset(train_features, label=train_target)

lgb_estimator = lightgbm.LGBMRegressor(boosting_type='gbdt')

gsearch = GridSearchCV(estimator=lgb_estimator, param_grid=param_grid, cv=5 , verbose=1)
lgb_model = gsearch.fit(X=train_features, y=train_target)

print(lgb_model.best_params_, lgb_model.best_score_)
```

Обучим модель с полученными параметрами и посчитаем RMSE

```python
train_data = lightgbm.Dataset(train_features, label=train_target)
parameters = {
        'application': 'regression',
        'metric': 'rmse',
        'boosting': 'gbdt',
        'max_bin': 600
    }
lightgbm_model = lightgbm.train(parameters,
                           train_data)
y_pred_dt = lightgbm_model.predict(train_features)
print("RMSE тренировочной выборки: ", mean_squared_error(y_pred_dt, train_target) ** 0.5)
```

# 4. Тестирование

Проверим полученные модели на тестовой выборке.

```python
def calculate_test(model, name, test_target, test_features):
    pred_test = model.predict(test_features)
    print("RMSE тестовой выборки для : ", name, mean_squared_error(pred_test, test_target) ** 0.5)
```

```python
test_target = test["num_orders"]
test_features = test.drop(columns=["num_orders"])
calculate_test(lin_reg_train, "LinearRegression", test_target, test_features)
calculate_test(dt_regressor, "DecisionTreeRegressor:", test_target, test_features)
calculate_test(lightgbm_model, "LightGBM", test_target, test_features)
```

Как оказалось наилучший результат показал LightGBM, и чуть хуже LinearRegression.  
DecisionTree справился хуже всего.  
Любопытен факт столь большой разницы между тестовым и тренировочным rmse. По всей видимости feature которые мы подобрали привели к сильному переобучению для наших моделей, особенно это видно для DecisionTree.
