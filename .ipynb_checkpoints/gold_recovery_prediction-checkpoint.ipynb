##### Проверьте, что эффективность обогащения рассчитана правильно. Вычислите её на обучающей выборке для признака rougher.output.recovery. Найдите MAE между вашими расчётами и значением признака. Опишите выводы.

Напишем функцию расчета значения rougher.output.recovery

Изменим функцию таким образом, чтобы она возвращала NaN вместо 0, чтобы не путать в дальнейшем нулевые истинно нулевые и ошибочные значения


```python
def recovery_efficiency_calc(gold_conc_after, gold_conc_before, gold_conc_tail):
    if (gold_conc_before == 0) | (gold_conc_after == gold_conc_tail):
        return np.NaN
    return gold_conc_after * (gold_conc_before - gold_conc_tail) / gold_conc_before / (gold_conc_after - gold_conc_tail) * 100
```

Добавим функцию расчета rougher.output.recovery в рамках датафрейма

```python
def rougher_output_calculation(df):
    rougher_output_recovery = []
    for index, row in df.iterrows():
        efficiency_score = recovery_efficiency_calc(row["rougher.output.concentrate_au"],
                                                       row["rougher.input.feed_au"], 
                                                       row["rougher.output.tail_au"])
        rougher_output_recovery.append(efficiency_score)
    rougher_output_recovery = pd.Series(rougher_output_recovery)
    return rougher_output_recovery
```

```python
rougher_output_recovery = rougher_output_calculation(train)
rougher_output_orig = train["rougher.output.recovery"]

rougher_output_recovery_nonNan = rougher_output_recovery.dropna()
rougher_output_orig_nonNan = rougher_output_orig.dropna()

common_index = rougher_output_orig_nonNan.index.intersection(rougher_output_orig_nonNan.index)

mean_absolute_error(rougher_output_recovery[common_index], rougher_output_orig[common_index])
```

##### 2.3. Исследуйте суммарную концентрацию всех веществ на разных стадиях: в сырье, в черновом и финальном концентратах. Заметили аномальные значения в суммарном распределении или нет? Если они есть, стоит ли их удалять из обеих выборок? Опишите выводы и удалите аномалии.

Выделим функцию sum_calculation. Функция будет принимать на вход датафрейм и проводить расчет суммы концетрации металов на трех разных стадиях. После получение итоговых данных функция построит гистограммы.

```python
def sum_calculation(df):
    sum_rougher_input_arr = []
    sum_rougher_output_arr = [] 
    sum_final_arr = []

    for index, row in df.iterrows():
        sum_rougher_input = row['rougher.input.feed_au'] + row['rougher.input.feed_ag'] + row['rougher.input.feed_pb']
        sum_rougher_output = row['rougher.output.concentrate_au'] + row['rougher.output.concentrate_ag'] + row['rougher.output.concentrate_pb']
        sum_final = row['final.output.concentrate_au'] + row['final.output.concentrate_ag'] + row['final.output.concentrate_pb']
        sum_rougher_input_arr.append(sum_rougher_input)
        sum_rougher_output_arr.append(sum_rougher_output)
        sum_final_arr.append(sum_final)
    frame = {'input': pd.Series(sum_rougher_input_arr), 
             'rougher_output': pd.Series(sum_rougher_output_arr), 
             'final': pd.Series(sum_final_arr) } 

    info = pd.DataFrame(frame) 
    display(info.plot(kind='hist', alpha=0.8, bins=30, figsize=[16,5]))
        
```

```python
sum_calculation(train)
```

Получим близкое к нормальному распределению для каждой из стадий. Однако видим большое количество нулевых значений, от которых необходимо избавится. \
Добавим функцию dataframe_trim которая удалит строки, в которых суммарная концентрация трех компонентов на любом из этапов меньше 10 для входных значений, 30 для результатов грубой обработки и 50 для финальных результатов.

```python
def dataframe_trim(df):
    df = df[(df['rougher.input.feed_au'] + df['rougher.input.feed_ag'] + df['rougher.input.feed_pb']) > 10]
    df = df[(df['rougher.output.concentrate_au'] + df['rougher.output.concentrate_ag'] + df['rougher.output.concentrate_pb']) > 30]
    df = df[(df['final.output.concentrate_au'] + df['final.output.concentrate_ag'] + df['final.output.concentrate_pb']) > 50]
    return df
```

##### 3.1. Напишите функцию для вычисления итоговой sMAPE.

```python
def smape(A, F):
    return 100/len(A) * np.sum(2 * np.abs(F - A) / (np.abs(A) + np.abs(F)))
```

##### 3.2. Обучите разные модели и оцените их качество кросс-валидацией. Выберите лучшую модель и проверьте её на тестовой выборке. Опишите выводы.

Создадим набор функций позволяющие выбрать наилучшую модель для расчетов а также возвращающую результат кросс-валидации и саму модель

```python
# Функция вычисления smape
def get_model_score(model, features_valid, target_valid):
    predict = model.predict(features_valid)
    score = smape(target_valid, predict)
    return score

# Обрезаем колонки у оригинального датафрейма по колонкам второго
def trim_feature_columns(df, df2):
    columns = df2.columns
    return df[columns]

# получим модель в зависимости от переданного типа
def get_model(model_type, feature_train, target_train):
    if (model_type == 'linear'):
        model = LinearRegression();
    elif (model_type == 'random_forest'):
        model = RandomForestRegressor(max_depth=14, n_estimators=40, random_state=12345);
    elif (model_type == 'decision_tree'):
        model = DecisionTreeRegressor(max_depth=14, random_state=12345);
    model.fit(feature_train, target_train)
    return model

# Получим среднюю оцену после проведения кросс-валидации
def calculate_model_score(model_type, data, target_column):
    scores = []
    block_size = int(np.ceil(len(data)/4))
    target = data[target_column]
    features = trim_feature_columns(data.drop(target_column, axis=1), test)
    for i in range(0, len(data), block_size):
        valid_indexes = data[i:].head(block_size).index
        train_indexes = data.drop(valid_indexes, axis=0).index
        
        features_train = features.loc[train_indexes]
        target_train = target.loc[train_indexes]
        features_valid = features.loc[valid_indexes]
        target_valid = target.loc[valid_indexes]

        model = get_model(model_type, features_train, target_train)
        score = get_model_score(model, features_valid, target_valid)
        scores.append(score)
    scores_mean = pd.Series(scores).mean()
    return scores_mean, model
```

Проведем вычисления для колонки "rougher.output.recovery"

```python
linear_scores_smape_rougher, linear_model_rougher_rougher = calculate_model_score('linear', train, "rougher.output.recovery")
random_forest_scores_smape_rougher, random_forest_model_rougher = calculate_model_score('random_forest', train, "rougher.output.recovery")
decision_tree_scores_smape_rougher, decision_tree_model_rougher = calculate_model_score('decision_tree', train, "rougher.output.recovery")
print("linear: ",linear_scores_smape_rougher, "random_forest:",random_forest_scores_smape_rougher,"decision_tree:",decision_tree_scores_smape_rougher)
```

```python
linear_scores_smape_final, linear_model_rougher_final = calculate_model_score('linear', train, "final.output.recovery")
random_forest_scores_smape_final, random_forest_model_final = calculate_model_score('random_forest', train, "final.output.recovery")
decision_tree_scores_smape_final, decision_tree_model_final = calculate_model_score('decision_tree', train, "final.output.recovery")
print("linear: ",linear_scores_smape_final,  "random_forest:", random_forest_scores_smape_final, "decision_tree:", decision_tree_scores_smape_rougher)
```

```python
final_smape = 0.25 * random_forest_scores_smape_rougher + 0.75 * random_forest_scores_smape_final
print(final_smape)
```

Финальный smape для тренировочных данных 9.2%
